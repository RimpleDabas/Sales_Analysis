# Import all the libraries

```python
import numpy as np
import pandas as pd       
import warnings
warnings.filterwarnings('ignore')
```

## read in the file and display

```python
df = pd.read_csv('Raw_data.csv')
df.head()
```

## check datatypes

```python
df.info()
```

## change the format of the columns 
we cannot do for postal code yet because in later steps we see there are null values

```python
df['Order Date'] = pd.to_datetime(df['Order Date'],format = "%d/%m/%Y")
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format = "%d/%m/%Y")
# df['Postal Code'] = df['Postal Code'].astype('int64')
df.info()
```

## check if there are any null values 
its true hence we have null values somewhere in the dataset

```python
df.isnull().values.any()
```

Find the null values

```python
df.isnull().sum()
```

get non-null values

```python
df = df[df['Postal Code'].notna()]
```

```python
df.isnull().sum()
```

```python
# now convert postal code
df['Postal Code'] = df['Postal Code'].astype('int64')
df.info()
```

```python
# check for the duplicated values
df_duplicate = df[df.iloc[:,1:].duplicated(keep=False)]
df_duplicate
```

```python
#check if ship date is before order that because tthis does not make sense if true
df[df['Ship Date'] < df['Order Date']]
```

```python
# find and drop the duplicates
df = df.iloc[:,1:].drop_duplicates()
df_duplicate = df[df.iloc[:,1:].duplicated(keep=False)]
df_duplicate
```

```python
# check if the column values are consistent
for each in list(df.columns):
    print(each,df[each].value_counts())
```

```python
# rename the columns
df.columns
```
